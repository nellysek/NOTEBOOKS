```python
import MySQLdb

db = MySQLdb.connect(user = "root",
                     passwd = "xxxx",
                     host = "localhost",
                     port = 3306,
                     use_unicode = True)
cursor = db.cursor()
```

# EXECUTION OF METRICS

```python
#Lists  issues table from opensrack code review (Reviews Data base)
cursor.execute(" SELECT * FROM openstack_reviews.issues LIMIT 10 ")

cursor.fetchall()
```

```python
#List changes tables from openstack code reviews(Reviews)
cursor.execute("SELECT * FROM openstack_reviews.changes LIMIT 10")

cursor.fetchall()
```

```python
#Total number of code reviews in OPenstack code reviews
cursor.execute("""select count(field) 
                  from openstack_reviews.changes 
                  where field='Code-Review'""")
cursor.fetchall()
```

```python
#Total number of issues in openstack code review
cursor.execute("""select count(distinct(issue_id)) 
                  from openstack_reviews.changes""")

cursor.fetchall()

# Alternatives from issues tables
cursor.execute("""select count(issue) 
                  from openstack_reviews.issues""")

cursor.fetchall()
```

```python
#Counts and categorises values assigned during code review
cursor.execute(""" select new_value, count(field)
                   from openstack_reviews.changes 
                   where field='Code-Review' group by new_value""")

cursor.fetchall()
```

```python
#Number of code review  patches by year and month
cursor.execute(""" select year(changed_on),   month(changed_on), new_value, count(field) as NumberOfCodeReview  
                   from openstack_reviews.changes 
                   where field='Code-Review'
                   group by year(changed_on), month(changed_on) 
                   order by year(changed_on) asc, month(changed_on) asc """)
cursor.fetchall()
```

```python
# Number of code review for patches assigned value -1 by year and month
cursor.execute(""" select year(changed_on), month(changed_on), new_value, count(field) 
                   from openstack_reviews.changes 
                   where field='Code-Review' and new_value=-1 
                   group by year(changed_on), month(changed_on) LIMIT 10 """)
cursor.fetchall()
               
```

```python
#Patchsets per issue with the gerrit Id(issue)
cursor.execute(""" select count(distinct(c.old_value)) as NumberOfPachsets, c.issue_id, i.issue
                   from openstack_reviews.changes c, openstack_reviews.issues i  
                   where c.issue_id=i.id and c.old_value!="" 
                   group by c.issue_id """)
cursor.fetchall()
```

```python
#Patchsets per issue with the gerrit Id(issue) using Inner join
cursor.execute(""" select count(distinct(c.old_value)) as NumberOfPachsets, c.issue_id, i.issue 
                   from openstack_reviews.changes c inner join openstack_reviews.issues i 
                   on c.issue_id=i.id
                   where c.old_value!=""
                   group by c.issue_id LIMIT 10 """)
cursor.fetchall()
```

```python
# number of people involved in the code review
cursor.execute(""" select count(distinct(changed_by))
                   from openstack_reviews.changes 
                   where field='Code-Review' """)
cursor.fetchall()
```

```python
#number of Developers submmiting Issues
cursor.execute(""" select count(distinct(submitted_by))
                   from openstack_reviews.issues """)
cursor.fetchall()
```

```python
#For every patches how many persons were involved
cursor.execute(""" select new_value, count(distinct(changed_by))
                   from openstack_reviews.changes
                   group by new_value """)
cursor.fetchall()

```

```python
#For each issue, the time of the first and the last event in changes
cursor.execute(""" select min(changed_on),max(changed_on), issue_id, issue
                   from openstack_reviews.changes c 
                   inner join openstack_reviews.issues i 
                   on c.issue_id=i.id
                   group by c.issue_id """)
cursor.fetchall()
```

```python
#For each issue, the difference between the first and last event in changes
cursor.execute(""" select  issue_id, TIMEDIFF(max(changed_on),min(changed_on))
                   from openstack_reviews.changes
                   group by issue_id LIMIT 20 """)
cursor.fetchall()
```
